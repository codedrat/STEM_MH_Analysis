```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mentalhealth_dataset.csv')
df.head()
```

```python
df.info()
```

```python
#standardizing string values
df_clean = df.copy()
df.YearOfStudy.unique()
```

```python

string_cols = ['Gender', 'Course','YearOfStudy']
for col in string_cols:
    df_clean[col] = df[col].str.lower().str.strip()
print(f"Standardized Years: {df_clean['YearOfStudy'].unique()}")
print(f"Standardized Majors: {df_clean['Course'].unique()}")
```

```python
df_clean.head()
```

```python
#Majors corrections and translations
course_cleanup = {
    'bit': 'it',
    'koe': 'engineering',
    'bcs': 'computer science',
    'irkhs': 'ahas kirkhs',
    'kirkhs': 'ahas kirkhs',
    'cts': 'computer science',
    'econs': 'economics',
    'malcom': 'communication',
    'diploma nursing': 'nursing',
    'benl': 'english learning and literature',
    'diploma tesl': 'teaching english as second language',
    'laws': 'law',
    'kenms': 'economics',
    'enm': 'economics',
    'engine': 'engineering',
    'engin': 'engineering',
    'mhsc': 'health science',
    'kop': 'pharmacy',
    'ala': 'arabic language and literature',
    'human sciences': 'human science',
    'taasl': 'teaching arabic as second language',
    'mathemathics': 'mathematics'
}
df_clean['Course'] = df_clean['Course'].replace(course_cleanup)
df_clean['Course'].value_counts()
```

```python
#making columns to ensure distribution for variance
def categorize_major(course):
    stem = ['engineering', 'computer science', 'it', 'biomedical science', 'nursing', 'marine science', 'mathematics', 'biotechnology', 'pharmacy', 'health science', 'radiography']
    humanities = ['psychology', 'communication', 'english learning and literature', 'human science', 'teaching english as second language', 'human resources', 'teaching arabic as second language', 'arabic language and literature']
    law_bus_econ = ['law', 'economics', 'business administration', 'accounting', 'banking studies']
    religion = ['ahas kirkhs', 'pendidikan islam', 'islamic education', 'fiqh', 'usuluddin', 'fiqh fatwa']

    if course in stem: return 'STEM'
    if course in humanities: return 'Humanities'
    if course in law_bus_econ: return 'Law/Bus/Econ'
    if course in religion: return 'Religious Studies'
    return 'Other'

df_clean['Major_Category'] = df_clean['Course'].apply(categorize_major)
print(df_clean['Major_Category'].value_counts())
```
